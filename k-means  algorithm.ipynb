```python
import numpy as np
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
```

```python
n_samples=500
```

```python
n_features=20
```

```python
n_clustres=4
```

```python
data ,_=make_blobs(n_samples=n_samples,centers=n_clustres,n_features=n_features,random_state=42)
```

```python
kmeans=KMeans(n_clusters=n_clustres,random_state=42)
kmeans.fit(data)
```

```python
labels = kmeans.predict(data)
```

```python
centroids = kmeans.cluster_centers_
```

```python
print("Cluster Centroids:\n", centroids)
print("\nCluster Labels for each data point:\n", labels)
```

```python
data
```

```python
plt.figure(figsize=(10,6))
plt.scatter(data[:,0],data[:,1], c=labels,  cmap='viridis', marker='o', s=50, alpha=1)
plt.scatter(centroids[:, 0], centroids[:, 1], c='red', marker='x', s=100, label='Centroids')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.title('K-Means Clustering Visualization (First Two Dimensions)')
plt.legend()
plt.show()
```
